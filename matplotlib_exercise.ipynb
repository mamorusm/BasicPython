## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df = pd.read_csv('data\hist_data.csv')
x1 = df['aデータ']
x2 = df['bデータ']

plt.hist(x1, bins=30, color='navajowhite', label='aデータ')
plt.hist(x2, bins=30, alpha=0.5, color='cyan', label='bデータ')
#色の微妙な違いが難しく、どの色を使うのが最適か分かりませんでした

plt.xlabel('温度(℃)')

plt.ylabel('度数')

plt.legend(loc='upper right')

plt.title('aデータとbデータの温度に関する度数分布')

plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df2 = pd.read_csv('data\scatter_data.csv')
x = df2['class']

condition0 = (x == 0)
condition1 = (x == 1)
condition2 = (x == 2)

n0 = df2[condition0]
n1 = df2[condition1]
n2 = df2[condition2]

x0 = n0['length (cm)']
y0 = n0['width (cm)']
x1 = n1['length (cm)']
y1 = n1['width (cm)']
x2 = n2['length (cm)']
y2 = n2['width (cm)']

plt.scatter(x0, y0, c='red', label='class0')
plt.scatter(x1, y1, c='blue',label='class1')
plt.scatter(x2, y2, c='green',label='class2')

plt.xlabel('length (cm)')
plt.ylabel('width (cm)')

plt.grid(which='both')

plt.legend(loc='upper right')

plt.title('クラス0～2のlengthとwidthに関する散布図')

plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df3 = pd.read_csv('data\line_data.csv')
x = df3['期間']
y = df3['件数']
plt.plot(x,y,marker='o')
plt.xlabel('期間')
plt.ylabel('件数')
plt.xticks(rotation=30)
plt.grid(True)
plt.title('期間ごとの件数の推移')
```

```python
x = df3['期間']
y = df3['費用']
plt.bar(x, y)
plt.xticks(rotation=30)
plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.grid(axis='both')
plt.title('期間ごとの費用の推移')
```
